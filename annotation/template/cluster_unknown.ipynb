```python
# import os
# from rich import inspect, print as rprint
from pathlib import Path
# import numpy as np
# import pandas as pd
import anndata as ad
import scanpy as sc

from spida.P.setup_adata import _calc_embeddings
# from spida.I.allcools import run_allcools_seurat

import matplotlib.pyplot as plt
import seaborn as sns

from spida.pl import plot_categorical
# Move Plot ALLCools functions outside of the cli
plt.rcParams['axes.facecolor'] = 'white'
```

```python
#parameters
EXPERIMENT = "PU"
label_col = "allcools_Subclass_filt"
```

```python
combo_path = f"/home/x-aklein2/projects/aklein/BICAN/data/aggregated/BICAN_BG_{EXPERIMENT}_comb_cellpose_SAM_filt.h5ad"
adata = ad.read_h5ad(combo_path)
adata
```

```python
adata_unk = adata[adata.obs[label_col] == "unknown"].copy()
adata_unk
```

```python
sc.pp.highly_variable_genes(adata_unk, n_top_genes=200, subset=True, flavor="seurat")
```

```python
adata_unk = _calc_embeddings(adata_unk, knn=25, leiden_res=0.5, layer="volume_norm", key_added="unk_")
```

```python
fig, ax = plt.subplots(1,3,figsize=(15,5))
plot_categorical(adata_unk, coord_base="unk_umap", cluster_col = "unk_leiden", show=False, coding=True, text_anno=True, ax=ax[0])
plot_categorical(adata_unk, coord_base="unk_umap", cluster_col = "donor", show=False, ax=ax[1])
plot_categorical(adata_unk, coord_base="unk_umap", cluster_col = "replicate", show=False, ax=ax[2])
plt.show()
```
